#1 Basic Data Exploration

```python
import pandas as pd

salaries_df = pd.read_csv('/content/drive/MyDrive/shAI/Salaries.csv')

num_rows, num_columns = salaries_df.shape
print(f"Number of rows: {num_rows} , Number of columns: {num_columns}")

data_types = salaries_df.dtypes
print("\nData types of each column:\n", data_types)

missing_values = salaries_df.isnull().sum()
print("\nMissing values in each column:\n", missing_values)
```

#2 Descriptive Statistics

```python
basic_stats = salaries_df[['TotalPay']].describe()

mode_salary = salaries_df['TotalPay'].mode()

salary_range = salaries_df['TotalPay'].max() - salaries_df['TotalPay'].min()

std_deviation = salaries_df['TotalPay'].std()

print("\nBasic Statistics:\n", basic_stats)
print("\nMode of Salary:", mode_salary.to_list())
print("\nRange of Salaries:", salary_range)
print("\nStandard Deviation of Salaries:", std_deviation)
```

#3 Data Cleaning

```python
# Check before data cleaning
print("Missing values before cleaning:\n", salaries_df.isnull().sum())

salaries_df['TotalPay'].fillna(salaries_df['TotalPay'].mean(), inplace=True)

# Check after data cleaning
print("\nMissing values after cleaning:\n", salaries_df.isnull().sum())
```

#4 Basic Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# to visualize the distribution of salaries (TotalPay)
plt.figure(figsize=(12, 6))
sns.histplot(salaries_df['TotalPay'], bins=30, kde=True, color='green')
plt.title('Distribution of Total Pay')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

# represent the proportion of employees in different departments
department_counts = salaries_df['JobTitle'].value_counts()
top_departments = department_counts.head(10)

plt.figure(figsize=(10, 10))
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Top 10 Departments')
plt.show()
```

#5 Grouped Analysis

```python
# Group data by 'Year'
grouped_by_year = salaries_df.groupby('Year')['TotalPay'].agg(['mean', 'median', 'min', 'max', 'std', 'count'])

# Display the grouped analysis
print("\nGrouped Analysis by Year:\n", grouped_by_year)
```

#6 Simple Correlation Analysis

```python
# Identify correlation between 'TotalPay' and 'Benefits'
correlation = salaries_df['TotalPay'].corr(salaries_df['Benefits'])
print("\nCorrelation between Total Pay and Benefits:", correlation)

# Visualize the relationship using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(salaries_df['TotalPay'], salaries_df['Benefits'], alpha=0.5)
plt.title('Scatter Plot: Total Pay vs. Benefits')
plt.xlabel('Total Pay')
plt.ylabel('Benefits')
plt.show()
```
